```python
# selecting random points 
# we want to have NxD matrix representing N poins in D dimensions.
```

```python
import numpy as np 
rng = np.random.default_rng(seed = 1701)
```

```python
# two dimensional normal distribution

mean = [0,0]
cov = [
    [1,2],
    [2,5]
]
X = rng.multivariate_normal(mean,cov,100)
X.shape 
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt 
plt.style.use("seaborn-whitegrid")

plt.scatter(X[:,0], X[:,1]);
```

```python
# let's with fancy indexing select 20 random points 
indices = np.random.choice(X.shape[0], 20, replace=False)
indices 
```

```python
selection = X[indices]
selection.shape 
```

```python
# add additional graphic into the plot 
plt.scatter(X[:,0],X[:,1], alpha=0.3)       # original points 
plt.scatter(selection[:,0], selection[:,1], facecolor="none", edgecolor="black",s=200)
```
