```python
# Simple Line Plots 

%matplotlib inline 
import matplotlib.pyplot as plt 
plt.style.use("seaborn-whitegrid")
import numpy as np 
```

```python
# two primitives we have: 
# figures (single container that contains axes, graphics, text, labels)
# axises (ticks, grids, labels)

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x));
```

```python
#       ,  plt.""
plt.plot(x, np.sin(x));
```

```python
#    plt.""     
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
# Adjusting colos and style
plt.plot(x, np.sin(x-1), color="blue", linestyle="solid")
plt.plot(x, np.sin(x-2), color="g", linestyle="dashed")
plt.plot(x, np.sin(x-3), color="#ff00ff", linestyle="dotted")
plt.plot(x, np.sin(x-4), color=(1.0,0.2,0.3))

plt.plot(x, np.sin(x-5), "-g")  # solid green
plt.plot(x, np.sin(x-6), ":r")  # dotted red 
```

```python
# Axes limits 
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
# auto tight the bounds 
plt.plot(x, np.sin(x))
plt.axis("tight")
```

```python
# axis ratio 
plt.plot(x, np.sin(x))
plt.axis("equal")
```

```python
# Labeling Plots 

plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
plt.plot(x, np.sin(x), "-g", label="sin(x)")
plt.plot(x, np.cos(x), ":b", label="cos(x)")
plt.axis("equal")

plt.legend()
```

```python
# comvetion from axis to plot itself

# plt.xlabel => ax.set_xlabel 
# plt.ylabel => ax.set_ylabel 
# plt.xlim => ax.set_xlim
# plt.ylim => ax.set_ylim
# plt.title => ax.set_title
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))

ax.set(xlim=(0, 10), ylim=(-2, 2),
        xlabel='x', ylabel='sin(x)',
        title='A Simple Plot')

plt.show()
```
