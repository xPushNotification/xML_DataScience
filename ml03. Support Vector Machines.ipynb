```python
# In Depth: Support Vector Machine
# опять же как обычно математика нам не особенно важна 
# по сути все что делает SVM это делит кластера с дополнительными векторами 
# которые задают серую зону - то есть пытается сделать так чтобы 
# в серой зоне было бы как можно меньше спорных элементов 

%matplotlib inline 
import numpy as np 
import matplotlib.pyplot as plt 
plt.style.use("seaborn-whitegrid")
from scipy import stats 
```

```python
from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples=50, centers=2, 
                 random_state=0, cluster_std=0.60)
plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap="autumn")
```

```python
# для начала покажем почему обычная классификация это плохо 
# попробуем просто разделить точки прямыми линиями 
# получив при этом мульти возможную линию 

xfit = np.linspace(-1, 3.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plt.plot([0.6], [2.1], 'x', color='red', markeredgewidth=2, markersize=10)
for m, b in [(1, 0.65), (0.5, 1.6), (-0.2, 2.9)]:
    plt.plot(xfit, m * xfit + b, '-k')
plt.xlim(-1, 3.5);
```

```python
# смысл SVM в том чтобы уменьшить как можно больше возможное нахождение 
# точек в серой зоне (на примере этого нет - так как тут двухмерный глупый случай)
# то есть проще говря: смысл в том - чтобы увеличить при разделении отступ 
# одного датасета от другого 

xfit = np.linspace(-1,3.5)
plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap="autumn")

for m,b,d in [(1, 0.65, 0.33), (0.5, 1.6, 0.55), (-0.2, 2.9, 0.2)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, "-k")
    plt.fill_between(xfit, yfit-d, yfit+d, edgecolor="none",
                     color="lightgray", alpha=0.5)
plt.xlim(-1, 3.5)

# очевидно что та линия в которой самая большая рамка (margin?) 
# собственно и лучше всего разделяет кластера 
```

```python
# fitting a support vector machine 

from sklearn.svm import SVC # support vector classifier
model = SVC(kernel="linear", C=1E10)
model.fit(X,y)
```

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
        levels=[-1, 0, 1], alpha=0.5,
        linestyles=['--', '-', '--'])
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
        model.support_vectors_[:, 1],
        s=300, linewidth=1, edgecolors='black',
        facecolors='none');

    ax.set_xlim(xlim)
    ax.set_ylim(ylim)


plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap="autumn")
plot_svc_decision_function(model);
```

```python
# Beyond Linear Boundaries: Kernel SVM 
# суть для чего нужно ядро - это добавить субсету еще Z ось для того чтобы 
# разрезы шли плоскостями - и точки нормально друг от друга 
# могли бы отделяться
# опять же то - как это считается - нас интересует меньше всего 
# по сути речь идет о том - чтобы свести все через PCA к трем измерениям 
# после чего с помощью классификатора на ядрах нарезать все плоскостями 
# так мы будем видеть результаты 

from sklearn.datasets import make_circles
X,y = make_circles(100, factor=.1, noise=.1)

clf = SVC(kernel="linear").fit(X,y)

plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap="autumn")
plot_svc_decision_function(clf, plot_support=False)
```

```python
# исправляем - перенося данные в более высокоуровневое разрешение
r = np.exp(-(X**2).sum(1))
```

```python
from mpl_toolkits import mplot3d

ax = plt.subplot(projection="3d")
ax.scatter3D(X[:,0], X[:,1], r, c=y, s=50, cmap="autumn")
ax.view_init(elev=20, azim=30)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("r")

# теперт на уровне 0.7 как видно можно нарисовать плоскость разделения 
# и кластеры точек будут разделены 
```

```python
X,y = make_blobs(n_samples=100, centers=2, 
                 random_state=0, cluster_std=0.8)
fig,ax = plt.subplots(1,2, figsize=(16,6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

for axi,C in zip(ax, [10.0, 0.1]):
    model = SVC(kernel="linear", C=C).fit(X,y)
    axi.scatter(X[:,0], X[:,1], c=y, s=50, cmap="autumn")
    plot_svc_decision_function(model, axi)
    axi.scatter(model.support_vectors_[:, 0],
    model.support_vectors_[:, 1],
    s=300, lw=1, facecolors='none');
    axi.set_title('C = {0:.1f}'.format(C), size=14)
```
